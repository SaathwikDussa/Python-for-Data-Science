#                             1)     SCATTER PLOT


```python
plt.figure(figsize=(8,8),dpi=200)
sns.scatterplot(x='GDP per capita', y='Score', data=df, alpha =0.8,palette = 'inferno')
plt.title('Gdp vs score')
```

#                                               2) HEAT MAP


```python
plt.figure(figsize=(6,6),dpi=100)
sns.heatmap(df.corr(),annot=True,cmap="twilight_shifted")
plt.title=('Happiness Report')
plt.show()
```
